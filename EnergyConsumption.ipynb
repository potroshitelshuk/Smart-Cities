## Paralleled k-means for each k

```python
import pyspark
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.feature import VectorAssembler
from pyspark.sql import SQLContext

from tqdm.notebook import tqdm

import warnings 
warnings.filterwarnings('ignore')
```

```python
## rewrite all in numpy vectors

def calc_silhoette(df_res):
    df_center = df_res.iloc[:, 1:97].mean(axis=0)
    K = len(df_res['prediction'].unique())
    
    inter_cluster_sum = 0
    intra_cluster_sum = 0
    
    for k in range(K):
        df_k = df_res[df_res['prediction'] == k].iloc[:, 1:97]
        df_k_center = df_k.mean(axis=0)
        
        inter_cluster_sum += np.sqrt(np.sum((df_k_center - df_center) ** 2))
        intra_cluster_sum += np.sqrt(((df_k - df_k_center) ** 2).sum(axis=1)).mean()
        
    inter_cluster = inter_cluster_sum / K
    intra_cluster = intra_cluster_sum / K
    
    bd_silhoette = (inter_cluster - intra_cluster) / (max(inter_cluster, intra_cluster))
    
    return bd_silhoette

def calc_dunn(df_res):
    df_center = df_res.iloc[:, 1:97].mean(axis=0)
    K = len(df_res['prediction'].unique())
    min_dist = np.inf
    max_dist = 0
    
    for k in range(K):
        df_k = df_res[df_res['prediction'] == k].iloc[:, 1:97]
        df_k_center = df_k.mean(axis=0)
        
        dist = np.sqrt(np.sum((df_k_center - df_center) ** 2))
        min_dist = min(dist, min_dist)
        
        values = df_k.values
        dst = np.sqrt(((df_k - df_k_center) ** 2).sum(axis=1)).max()
        max_dist = max(max_dist, dst)
                
    return min_dist / max_dist

def calc_davies_bouldin(df_res):
    res = 0
    K = len(df_res['prediction'].unique())
    for i in range(K):
        max_value = 0
        for j in range(K):
            if i == j:
                continue
                
            df_i = df_res[df_res['prediction'] == i].iloc[:, 1:97]
            df_i_center = df_i.mean(axis=0)
            df_j = df_res[df_res['prediction'] == j].iloc[:, 1:97]
            df_j_center = df_j.mean(axis=0)
            
            r_i = np.sqrt(((df_i - df_i_center) ** 2).sum(axis=1)).mean()
            r_j = np.sqrt(((df_j - df_j_center) ** 2).sum(axis=1)).mean()
            dst = np.sqrt(((df_i_center - df_j_center) ** 2).sum())
            
            value = (r_i + r_j) / dst
            max_value = max(value, max_value)
        res += max_value
        
    return res / K
    
def calc_wssse(df_res):
    res = 0
    K = len(df_res['prediction'].unique())
    for k in range(K):
        df_k = df_res[df_res['prediction'] == k].iloc[:, 1:97]
        df_k_center = df_k.mean(axis=0)
        res += ((df_k - df_k_center) ** 2).sum(axis=1).sum()
    return res
            

def train(spark_data, k):
    res = {'n_clusters': k}
    
    # Trains a k-means model.
    kmeans = KMeans().setK(k).setSeed(1)
    model = kmeans.fit(spark_data)

    # Make predictions
    predictions = model.transform(spark_data)

    # Add predictions to data
    transformed = model.transform(spark_data).select('id', 'prediction')
    rows = transformed.collect()
    df_pred = SQLContext(sc).createDataFrame(rows)
    df_pred = spark_data.join(df_pred, 'id')
    pddf_pred = df_pred.toPandas().set_index('id')
    df_res = df_grouped.merge(pddf_pred['prediction'], on='id')

#     df_res = spark_data.copy(deep=True)
#     model = sklearn_kmeans(k)
#     model.fit(spark_data)
#     df_res['prediction'] = model.labels_

    # get score
    
    bd_silhoette_score = calc_silhoette(df_res)
    bd_dunn_score = calc_dunn(df_res)
    davies_bouldin_score = calc_davies_bouldin(df_res)
    wssse_score = calc_wssse(df_res)
    
    res['silhoette'] = bd_silhoette_score
    res['dunn'] = bd_dunn_score
    res['davies_bouldin'] = davies_bouldin_score
    res['wssse'] = wssse_score
    res['prediction'] = df_res['prediction']
    
    return res
```

```python
def plot_metrics(results):
    fig, ax = plt.subplots(2, 2, figsize=(16, 16))
    for i in range(4):
        ax[i // 2][i % 2].plot(results['n_clusters'].values, results.iloc[:, i + 1].values)
        ax[i // 2][i % 2].set_title(results.iloc[:, i + 1].name)
        ax[i // 2][i % 2].set_xlabel('K Values')
        ax[i // 2][i % 2].grid()
        
```

```python
def find_local_max(arr, clusters):
    local_maxs = []
    for i in range(2, len(arr)):
        if arr[i - 2] < arr[i - 1] and arr[i] < arr[i - 1]:
            local_maxs.append(clusters[i - 1])
        
    return local_maxs
```

```python
def add_features(df_grouped, df_idxs):
    df_grouped['season'] = None
    
    df_grouped.loc[df_grouped['month'].isin([12, 1, 2]), 'season'] = 'Winter'
    df_grouped.loc[df_grouped['month'].isin([3, 4, 5]), 'season'] = 'Spring'
    df_grouped.loc[df_grouped['month'].isin([6, 7, 8]), 'season'] = 'Summer'
    df_grouped.loc[df_grouped['month'].isin([9, 10, 11]), 'season'] = 'Autumn'
    
    df_grouped['weekday'] = pd.to_datetime(
        df_grouped['year'].astype(str) + '-' +
        df_grouped['month'].astype(str) + '-' + df_grouped['day'].astype(str)).dt.weekday
    
    df_grouped['weekday'] = df_grouped['weekday'].replace({
        0 : 'Monday',
        1 : 'Tuesday',
        2 : 'Wednesday',
        3 : 'Thursday',
        4 : 'Friday',
        5 : 'Day-off',
        6 : 'Day-off'
    })
    
    
add_features(df_grouped, df_idxs)
```

```python
def plot_chart(df_grouped, index, columns, values, ax, kind):
    pivot = pd.pivot_table(df_grouped, index=index, columns=columns, values=values, aggfunc='count')
    pivot = pivot.fillna(0)
    pivot = pivot.div(pivot.sum(axis=1), axis=0).mul(100)

    pivot = pivot
    cumulative = pivot.cumsum(axis=1)

    n_rows, n_cols = pivot.shape
    y_pos = range(n_rows)
    height = 0.8

    for i in range(n_cols):
        left = cumulative.iloc[:, i]-pivot.iloc[:, i]
        labels = []
        for value in pivot.iloc[:, i]:
            if value != 0:
                labels.append(f'{value:.1f}%')
            else:
                labels.append('')
        if kind == 'barh':
            ploted = ax.barh(y_pos, pivot.iloc[:, i], height, 
                             align='center', 
                             left=left, 
                             zorder=2,
                             label=pivot.columns[i])
        elif kind == 'bar':
            ploted = ax.bar(y_pos, pivot.iloc[:, i], height, 
                             align='center', 
                             bottom=left, 
                             zorder=2,
                             label=pivot.columns[i])
        ax.bar_label(ploted, label_type='center', fontsize=12, labels=labels)

    if kind == 'barh':
        ax.set_yticks(y_pos)
        ax.set_yticklabels(pivot.index)
        ax.tick_params(axis='y', pad=20)
        ax.set_xticks(range(0, 101, 10))
        ax.grid(axis='x', zorder=0)
        ax.legend()
    elif kind == 'bar':
        ax.set_xticks(y_pos)
        ax.set_xticklabels(pivot.index)
        ax.tick_params(axis='x', pad=20)
        ax.set_yticks(range(0, 101, 10))
        ax.grid(axis='y', zorder=0)
        ax.legend()
```

```python
fig, ax = plt.subplots(3, 2, figsize=(16, 30))
for i, feature in enumerate(['building', 'season', 'weekday']):
    ax[i][0].set_title(feature)
    ax[i][1].set_title(feature)
    plot_chart(df_grouped, 'prediction', feature, 'id', ax[i][0], 'barh')
    plot_chart(df_grouped, feature, 'prediction', 'id', ax[i][1], 'bar')
```
